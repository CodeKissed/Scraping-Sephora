```python
#preparation
## all imports
import requests
import re # regular expressions
import bs4 #this is beautiful soup
import time

import json
import numpy as np
import urllib.request
```

```python
#open brandlist website html as string
byteurl = requests.get("https://www.sephora.com/brands-list").content
brandlist = str(byteurl,'utf-8')
print(brandlist)
```

```python
# scraping each brand website
brand_url = re.findall(R'css-d84rnc.*?href=\"(.*?)\"',brandlist)
for line in brand_url[1:]:
    print("https://www.sephora.com"+line+"/all?pageSize=300")
```

```python
#Save these websites into a list
brand_link_lst = []
for line in brand_url[1:]:
    brand_link_lst.append("https://www.sephora.com"+line+"/all?pageSize=300")
```

```python
#How many brands? 
count1=0
for webb in brand_link_lst:
    count1 = count1+1
print("Brands #",count1) 

## which indicates how many requests we need to do to scape all the product links.
## The result shows 300+ brands. Normally, this will not exceed the request times that Sephora allows per day. 
## Therefore, we could scrape all the product links at one time.
```

```python
# Test: scrape product links and save them into a list
products = []
count2 = 0
for webb in brand_link_lst[:10]: ##Just use the first 10 brands for test.
    print(webb)
    brand_page_content = str(requests.get(webb).content,'utf-8')
    scrape_begin = brand_page_content.find('"products"')+11
    scrape_end = brand_page_content.find('"pageSize"')-1
    products_para = brand_page_content[scrape_begin:scrape_end]
    products_url = re.findall(R'targetUrl\"\:\"(.*?)\"',products_para)
    for product_link_name in products_url:
        print(product_link_name)
        if product_link_name.startswith('/product/') is True:
            products.append("https://www.sephora.com"+product_link_name)
            count2 += 1
    print(webb +" product #" + str(count2))
    count2 = 0
    time.sleep(np.random.random()*3)
count3 = 0
for item in products:
    count3 +=1
print('Total products #:',count3)    
```

```python
# Test: scrape product links and save them into a list
for item in products:
    print(item)
```

```python
#Formal: scrape product links and save them into a list
products = []
count2 = 0
for webb in brand_link_lst[:]: ## Unlike before, we scrape all the products at this time.
    brand_page_content = str(requests.get(webb).content,'utf-8')
    scrape_begin = brand_page_content.find('"products"')+11
    scrape_end = brand_page_content.find('"pageSize"')-1
    products_para = brand_page_content[scrape_begin:scrape_end]
    products_url = re.findall(R'targetUrl\"\:\"(.*?)\"',products_para)
    for product_link_name in products_url:
        products.append("https://www.sephora.com"+product_link_name)
        count2 += 1
    print(webb +" product #" + str(count2))
    count2 = 0
    time.sleep(np.random.random()*3)
count3 = 0
for item in products:
    count3 +=1
print('Total products #:',count3)

#write product link into a file:
with open('product_link.txt', 'w') as f:
    for item in products:
        f.write("%s\n" % item)
```
