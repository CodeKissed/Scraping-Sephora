```python
import pandas as pd
import re
```

```python
pd_data = pd.read_csv('../codes/data/product_data.csv')
pd_data_dropna = pd_data.dropna(subset=['Name'])
pd_data_dropna.info()
```

```python
pd_info = pd.read_csv('../codes/data/pd_info.csv')
pd_info.info()
```

```python
pd_info_dropna = pd_info.dropna(subset=['pd_id', 'size_and_item', 'category'])
pd_info_fillna = pd_info_dropna.fillna(0)
pd_info_fillna.info()
```

```python
pd_whole = pd.merge(pd_data_dropna, pd_info_fillna, on='pd_id')
pd_whole.info()
```

```python
pd_whole.sample(5)
```

From the sample dataframe and dataframe info, we can recognize that `Product_Id`, `item_num` and `link` may be not useful for analysis. Therefore, we are going to drop these three columns. Besides, `category` should be splited into three columns. `size` could have na value or other invalid value. Other columns like `love_count`, `rating`, `reviews_count` should turn to numeric variables.

# Let's start cleaning data!

## drop columns

```python
raw_data_c = pd_whole.drop(columns=['pd_id', 'product_links'])
raw_data_c.sample(5)
```

## Price

```python
raw_data_c[['AverageOverallRating', 'Age_13to17', 
            'Age_18to24', 'Age_25to34', 
            'Age_35to44', 'Age_45to54', 
            'Age_over54']] = raw_data_c[['AverageOverallRating', 'Age_13to17', 
                                         'Age_18to24', 'Age_25to34', 
                                         'Age_35to44', 'Age_45to54', 
                                         'Age_over54']].fillna(0)
```

```python
raw_data_c.info()
```

From the dataframe above, we can also observe that price may contain dollar sign as well. So we need to strip that.

```python
# remove dollar sign for price column
special_price = raw_data_c['price']
special_price_sr = special_price.str.extractall(r"[$](?P<number>\d+)")
special_price_df = special_price_sr.reset_index().pivot(
    index='level_0', columns='match', values='number').reset_index()
special_price_df.columns = ['index', 'Price_0', 'Price_1', 'Price_2']
special_price_df[['Price_0', 'Price_1', 'Price_2']] = special_price_df[[
    'Price_0', 'Price_1', 'Price_2']].astype(float)
```

```python
Price = special_price_df.drop(columns=['index']).max(axis=1)
Price.name = 'Price'
```

```python
add_price_df = pd.concat([raw_data_c, Price], axis=1)
clean_price = add_price_df.drop(columns=['price'])
clean_price.info()
```

```python
clean_price['FirstSubmissionTime'] = pd.to_datetime(clean_price['FirstSubmissionTime'], 
                                                    format='%Y-%m-%dT%H:%M:%S.000+00:00')
clean_price['LastSubmissionTime'] = pd.to_datetime(clean_price['LastSubmissionTime'], 
                                                    format='%Y-%m-%dT%H:%M:%S.000+00:00')
clean_price.info()
```

## split category into three columns

```python
category_lst_series = clean_price['category'].str.split(',')
clean_price['category_1'] = category_lst_series.str[0]
clean_price['category_2'] = category_lst_series.str[1]
clean_price['category_3'] = category_lst_series.str[2]
clean_price.sample(5)
```

## Size

As there are different measure for different products. Here, we mainly focus on products that can be measured by 'ml', 'oz', and 'g'.

Use regex to find measures with oz or mL or g and create responding columns

```python
def find_measure(measure, text):
    regex = '(\d?\.?\d+).?' + measure   
    result = re.findall(regex, text)
    if len(result)!=0:
        return result[0]
    else:
        return None
```

```python
measures = [' oz', ' mL', ' g']
for measure in measures:
    measure_ = measure.split(' ')[-1]
    name = 'size_' + measure_
    clean_price[name] = clean_price.loc[(clean_price['size_and_item'].str.contains(measure_)) &
                                          (~clean_price['size_and_item'].str.contains('x')), 'size_and_item'].apply(lambda x: find_measure(measure_, x))
clean_price.sample(5)
```

```python
clean_df = clean_price.drop(columns=['size_and_item'])
clean_df.info()
```

```python
clean_df.to_csv('cleaned_data.csv', index=False)
```
