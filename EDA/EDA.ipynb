```python
import pandas as pd
```

```python
df = pd.read_csv('../clean_data/cleaned_data.csv')
df.sample(5)
```

```python
df.info()
```

# Top 10 most expensive products

```python
df.sort_values('price', ascending=False).head(10)
```

# Top 10 most popular products

```python
df.sort_values('love_count', ascending=False).head(10)
```

# Factors on Price

## Brand effect on price

### Top 10 most expensive brand (by mean price)

```python
df.groupby('brand').mean().sort_values('price',ascending= False).head(10)
```

### Top 10 most expensive brand (by median price)

```python
df.groupby('brand').median().sort_values('price',ascending= False).head(10)
```

## Category effect on Price

### Mean price by main category

```python
df.groupby('category_1').mean().sort_values('price',ascending= False).head(10)
```

### check std

```python
df.groupby('category_1').std().sort_values('price',ascending= False).head(10)
```

### by sub category

```python
df.groupby('category_3').mean().sort_values('price',ascending= False).head(10)
```

# Brand effect on love counts

### Top 10 most popular brand (by mean)

```python
df.groupby('brand').mean().sort_values('love_count',ascending= False).head(10)
```

# Brand effect on ratings

### Top 10 most popular brand (by mean)

```python
by_brand = df.groupby('brand').mean()
by_brand.loc[by_brand.reviews_count >=30].sort_values('rating',ascending= False).head(10)
```
